```python
from pdf2image import convert_from_path
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path
import re
```

```python
files_path = f'/Users/pablo/Documents/GitHub/NNCalderon/reduced_data_checkerboard_trig_wavelet_op_lin_parab_and_op_lin_ricker_54BC/test_1/'
save_path = files_path + 'gifs/'

if os.path.isdir(save_path):
    print("Saving directory exists")
else:
    os.mkdir(save_path) 
    print("Saving directory created") 
```

```python

# === Settings ===
folder = Path(files_path)  # folder containing your PDFs
pattern = 'gamma_recovered_epoch_*.pdf'
output_gif = 'gamma_full_training.gif'
duration_ms = 500  # ms per frame

# Optional: choose a font
try:
    font = ImageFont.truetype("/Library/Fonts/Arial.ttf", 40)
except:
    font = ImageFont.load_default()

# === Collect and sort files ===
pdf_files = sorted(folder.glob(pattern), key=lambda f: int(f.stem.split('_')[-1]))

frames = []
for pdf_file in pdf_files:
    print(f"Converting {pdf_file} ...")

    # Extract epoch number
    match = re.search(r'(\d+)$', pdf_file.stem)
    epoch = match.group(1) if match else "?"

    # Convert first page of the PDF to an image
    images = convert_from_path(pdf_file, dpi=150)
    img = images[0].convert('RGBA')

    # Draw the epoch number on the image
    draw = ImageDraw.Draw(img)
    text = f"Epoch: {epoch}"

    # Get text bounding box
    bbox = draw.textbbox((0, 0), text, font=font)
    text_width = bbox[2] - bbox[0]
    text_height = bbox[3] - bbox[1]

    # Center text at the top
    x = (img.width - text_width) // 2
    y = 20

    # Draw a semi-transparent white box behind text
    box_padding = 10
    draw.rectangle(
        [x - box_padding, y - box_padding,
         x + text_width + box_padding, y + text_height + box_padding],
        fill=(255, 255, 255, 200)
    )

    # Draw the text
    draw.text((x, y), text, fill="black", font=font)

    frames.append(img)

# === Save as GIF ===
if frames:
    frames[0].save(
        save_path + output_gif,
        save_all=True,
        append_images=frames[1:],
        duration=duration_ms,
        loop=0
    )
    print(f"✅ GIF saved as {output_gif}")
else:
    print("❌ No matching PDF files found.")
```
